```python
import numpy as np
import pandas as pd
```

```python
record_1867 = pd.read_excel("final_1867.xlsx") # this is the excel sheet from the client's email titled 'final spreadsheet'
record_1867
```

```python
record_1867 = record_1867.drop(['Unnamed: 54', 'Unnamed: 55',
       'Unnamed: 56', 'Unnamed: 57', 'Unnamed: 58', 'Unnamed: 59'], axis=1)
```

```python
record_1867['EventLocJurisdictionCounty'].value_counts()
```

```python
record_1867 = record_1867.replace('Louisa ', 'Louisa')
record_1867['EventLocJurisdictionCounty'].value_counts()
```

```python
record_1867
```

```python
string_col = ['SourceSteward', 'SourceLocCity', 'SourceLocState', 'SourceTitle',
               'SourceType', 'SourceDateYearCreated', 'SourceCreator',
               'SourceLocCreatedCounty', 'SourceAuthorName', 'EventTitle',
               'EventLocJurisdictionCounty', 'EventDateYear', 'EventTranscriberNotes',
               'PersonSurname', 'PersonGivenNames', 'PersonNameAlternate',
               'PersonNameSuffix', 'PersonEventRole', 'PersonFN', 'PersonRoleLocSurnameEmployer', 
                'PersonRoleGivenNamesEmployer',
               'PersonAlternateNameEmployer', 'PersonRoleNameSuffixEmployer',
               'PersonRoleLocEmployment', 'PersonRoleLocResidence','PersonTaxCommissionerRemarks', 'EventImageLink']
```

```python
record_1867 = record_1867.replace(['', ' ', '-', '####', '`', '190x170', '..27', 'NaN', '55 mill', np.NaN], np.nan)

for col in list(record_1867.columns):
    
    if col in string_col:
        record_1867[col] = record_1867[col].fillna("")
        record_1867[col] = record_1867[col].astype(str)
    else:
        record_1867[col] = record_1867[col].fillna(0)
        record_1867[col] = record_1867[col].astype(float)
```
