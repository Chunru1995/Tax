## Name Search

```python
name = q1.name_query("George")
```

```python
len(name)
```

```python
name[0][["Child\nFirst Name", "Child\nLast Name", "Name of Place", 
        "Father/\nOwner\nFirst Name", "Father/\nOwner\nLast Name",
        "Mother\nFirst Name", "Mother\nLast Name", "First Name \nInformation\nProvided by",
        "Info Provider\nRelationship to\nPerson Born"]]
```

```python
location = q1.location_query("louisa")
```
